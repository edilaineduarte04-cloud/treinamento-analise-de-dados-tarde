Criar duas listas:
- 5 produtos (apenas inserir valores)
- A primeira lista, o custo da venda.
- A segunda lista, o preço de venda.
Desafio: Calcular o lucro de cada produto.

```python
preco_custo = [13, 8, 15, 12, 5]
custo_venda = [25, 14, 15, 16, 5]
array_quantidade = np.array(preco_custo)
array_custo_venda = np.array(custo_venda)

custo = array_quantidade - array_custo_venda
print (custo)
```

```python
salarios = np.array ([2,3,4,5,6,500])
maior_salario = np.max(salarios)
print(maior_salario)
```

```python
menor_salario = np.min(salarios)
print(menor_salario)
```

```python
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana_salarios = np.median(salarios)
print(mediana_salarios)
```

```python
desvio_padrao = np.std(salarios)
print (desvio_padrao)
```
